```python
#Importing pandas library for reading csv file and matplotlib for Visualising
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading the data
dt=pd.read_csv('http://bit.ly/w-data')
print(dt)
```

```python
#Visualising data
dt.plot(x='Hours',y='Scores',style='o',label='Scores')
plt.xlabel('Study Hours')
plt.ylabel('Students Score')
plt.title('Houres vs Scores')
plt.legend()
plt.show()
```

```python
#splitting the features and label
X=dt[['Hours']]
y=dt['Scores']
```

```python
#Providing training dataset and testing dataset
train_X=X[:20]
train_y=y[:20]
test_X=X[20:]
test_y=y[20:]
```

```python
#Visualising tarining dataset
plt.scatter(train_X,train_y,color='green')
plt.xlabel('Study Houres')
plt.ylabel('Students Score')
plt.title('Houres vs Scores of training dataset')
plt.show()
```

```python
#Training the Linear Regression model
from sklearn.linear_model import LinearRegression
from sklearn import metrics
l_reg=LinearRegression()
l_reg.fit(train_X,train_y)
print("Training Completed")
```

```python
#plotting LinearRegression line
#Y=mx+b
lr_line = l_reg.coef_*train_X+l_reg.intercept_
plt.scatter(train_X,train_y,color='red')
plt.plot(train_X,lr_line)
plt.xlabel('Study Houres')
plt.ylabel('Students Score')
plt.show()
```

```python
#Preditction of the Scores
prediction=l_reg.predict(test_X)
df=pd.DataFrame({'Predicted':prediction,'Actual':test_y})
print(df)
```

```python
#Visualisation for difference between predicted value and actual value
plt.scatter(test_X,test_y,label='Actual',color='green')
plt.scatter(test_X,prediction,label='Predicted',color='blue')
plt.title("Difference between Predicted value and actual value")
plt.xlabel('Study Houres')
plt.ylabel('Students Score')
plt.legend()
plt.show()
```

```python
#Predtion for who studied 9.25 Hours
hours=9.25
predict_score=l_reg.predict([[hours]])
print("Student studied for {} Hours".format(hours))
print("Predicted Score of that Student is {}".format(predict_score[0]))
```

```python
#Accuray of our model
acc=l_reg.score(test_X,test_y)
print("Accuray of our model is : {}%  ".format(acc*100))
```
